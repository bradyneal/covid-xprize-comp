```python
import matplotlib.pyplot as plt
%matplotlib inline
import neat
import numpy as np
```

```python
NUM_CHECKPOINTS = 100 # Set based on how many checkpoints have been saved so far.
```

```python
pop_arr = []
for i in range(NUM_CHECKPOINTS):
    #print('Checkpoint:', i)
    checkpoint = neat.Checkpointer.restore_checkpoint('neat-checkpoint-{}'.format(i))
    pop = checkpoint.population
    pop_arr.append(pop)
```

```python
x_arr = []
y_arr = []
for pos in range(0, len(pop_arr)-1):
    new_vals_x = [pop_arr[pos][key].fitness_mult[0] for key in pop_arr[pos].keys()]
    new_vals_y = [pop_arr[pos][key].fitness_mult[1] for key in pop_arr[pos].keys()]
    x_arr.extend(new_vals_x)
    y_arr.extend(new_vals_y)

plt.figure(figsize=(15,8))
plt.scatter(x_arr, y_arr, s=1)
pos=-1
x_arr = [pop_arr[pos][key].fitness_mult[0] for key in pop_arr[pos].keys()]
y_arr = [pop_arr[pos][key].fitness_mult[1] for key in pop_arr[pos].keys()]
plt.scatter(x_arr, y_arr, s=3, c='r')
plt.xlim((0,4))
plt.ylim((0,0.45))
plt.title('All solutions, final generation is in red')
```
